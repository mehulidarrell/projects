```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
d=pd.read_csv("/kaggle/input/simple-linear-regression-placement-data/placement.csv")
```

```python
plt.scatter(d['cgpa'],d['placement_exam_marks'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
X=d[['cgpa']]
Y=d['placement_exam_marks']
```

```python
model=LinearRegression()
model.fit(X,Y)
slope=model.coef_[0]
intercept=model.intercept_ 

```

```python
def linear_regression(x):
    slope*x + intercept
```

```python
plt.scatter(d['cgpa'],d['placement_exam_marks'],label='Data Points')
plt.plot(d['cgpa'], [linear_regression(x1) for x1 in d['cgpa']], color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression')
plt.show()
```

```python
n=8
prediction=linear_regression(n)
print("Prediction for x= (n): (prediction)")

n=pd.DataFrame({'cgpa':[8]})
print(model.predict(n))
```
